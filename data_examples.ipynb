```python
from data import *
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
%matplotlib inline
```

```python
def show_examples(images,n=3):
  fig = plt.figure(figsize=(3.*n, 3.*n))
  grid = ImageGrid(fig, 111,
                  nrows_ncols=(n, n),
                  axes_pad=0.1,
                  )
  for ax, im in zip(grid, [images[i] for i in range(9)]):
        ax.imshow(np.clip(im,a_min=0,a_max=1))
  plt.show()
```

```python
for placement in [Placement.random_loc, Placement.random_corner, Placement.fixed_corner]:
    train_xs, _, _, _ = load_data(dataset=Dataset.cifar10, image_size=16, background_size=32,
                                  image_noise=ImageNoise.none, background_noise=BackgroundNoise.gaussian,
                                  background_noise_scalar=0.5,
                                  placement=placement)
    show_examples(train_xs,n=2)
```

```python
train_xs, _, _, _ = load_data(dataset=Dataset.cifar4, image_size=16, background_size=32,
                              image_noise=ImageNoise.none, background_noise=BackgroundNoise.blocks,
                              block_noise_data=BlockNoiseDataset.cifar6,
                              background_noise_scalar=0.5)

show_examples(train_xs)
```
